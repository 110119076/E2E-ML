```python
def updateTeamNames(df,oldTeamName,newTeamName):
    df["team1"] = df["team1"].str.replace(oldTeamName, newTeamName)
    df["team2"] = df["team2"].str.replace(oldTeamName, newTeamName)
    df["toss_winner"] = df["toss_winner"].str.replace(oldTeamName, newTeamName)
    df["winner"] = df["winner"].str.replace(oldTeamName, newTeamName)
```

```python
updateTeamNames(match_totalRuns1, "Deccan Chargers", "Sunrisers Hyderabad")
updateTeamNames(match_totalRuns1, "Gujarat Lions", "Gujarat Titans")
updateTeamNames(match_totalRuns1, "Rising Pune Supergiant", "Chennai Super Kings")
updateTeamNames(match_totalRuns1, "Chennai Super Kingss", "Chennai Super Kings")
updateTeamNames(match_totalRuns1, "Delhi Daredevils", "Delhi Capitals")
updateTeamNames(match_totalRuns1, "Kings XI Punjab", "Punjab Kings")
```
